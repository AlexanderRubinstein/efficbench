```python
results = 'acc'
agg_metric = 'scenarios'

if results == 'acc': ylim = (0,.1)
elif results == 'rank': 
    if agg_metric == 'std': ylim = (0,.1)
    else: ylim = (.5,1)
else: raise NotImplementedError
    
table_avg = {}
table_std = {}
model_perf = {}
for bench in ['helm_lite','lb']:
    table_avg[bench] = {}
    table_std[bench] = {}
    model_perf[bench] = {}
        
    for split in splits[bench]:
        table_avg[bench][split] = {}
        table_std[bench][split] = {}
        model_perf[bench][split] = {}
        

        with open(f'results/results_{bench}_split-{split}_iterations-5.pickle', 'rb') as handle:
            data = pickle.load(handle)

        models = list(data.keys())
        number_items = list(data[models[0]].keys())
        scenarios = list(data[models[0]][number_items[0]][methods[0]].keys())
        data = np.array([[[[data[model][number_item][method][scenario] for scenario in scenarios]  for model in data.keys()] for number_item in number_items] for method in methods])
        data = data.mean(-1) #iterations
 
        for k,scenario in enumerate(scenarios):
            table_avg[bench][split][scenario] = {}
            table_std[bench][split][scenario] = {}
            
            for i,method in enumerate(methods):
                table_avg[bench][split][scenario][method] = {}
                table_std[bench][split][scenario][method] = {}

                for j,number_item in enumerate(number_items):
                    table_avg[bench][split][scenario][method][number_item] = np.mean(data[:,:,:,k], axis=-1)[i,j]
                    table_std[bench][split][scenario][method][number_item] = data[:,:,:,k].std(-1)[i,j]
```

```python
for bench in ['lb','helm_lite']:
    scenarios = list(table_avg[bench]['iid'].keys())

    for i, scenario in enumerate(scenarios):  # Replace `benchmarks` with your list of benchmarks

        plt.figure(figsize=(8,3))
        plt.subplots_adjust(wspace=.5, hspace=.25)

        for j, split in enumerate(splits[bench]):
            if bench=='helm':ylabel_dic = {'iid':'(random split)', 'noniid':'(organization shift)'}
            else: ylabel_dic = {'iid':'(random split)', 'noniid':'(date shift)'}
            if results == 'rank': ylabel=f'Spearman correlation \n for LLMs ranking\n{ylabel_dic[split]}'
            else: ylabel = f'performance estimation error\n{ylabel_dic[split]}'

            # Iterate over your benchmarks
            xlabel='number of examples (per scenario)'    
            # Plot on the ith subplot
            plt.subplot(1, 2, j+1)
            plot_perf_lines(table_avg[bench][split][scenario], table_std[bench][split][scenario],
                            title='',
                            xlabel=xlabel,
                            ylabel=ylabel, ylim=ylim, legend=j==1,
                            error_bar=True,show_title=True, show_xlabel=True, show_ylabel=True, posic=(-.275, -.35))#ylims[bench]
        plt.savefig(f'plots/{bench}_performance_{scenario}.png', bbox_inches='tight', dpi=400, transparent=True)
        plt.show()
```
